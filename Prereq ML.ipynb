# Q1. How many multiplications and additions do you need to perform a matrix multiplication between a (n, k) and (k, m) matrix? Explain.

There are k elements in the first row of the (n, k) matrix and k elements in the first column of the (k, m) matrix. Thus, k multiplications and k-1 additions will form one element of the resultant matrix. The resultant matrix is (n, m) and thus has n*m elements. So, in the matrix multiplication, k*m*n multiplications and (k-1)*m*n additions need to be performed.

# Q3. Finding the highest element in a list requires one pass of the array. Finding the second highest element requires 2 passes of the the array. Using this method, what is the time complexity of finding the median of the array? Can you suggest a better method? Can you implement both these methods in Python and compare against numpy.median routine in terms of time?

For finding the highest number, we need one pass, so, the time is O(n). Then, for the second highest element, the time complexity is O(2n). So for median, we need n/2 passes. Thus, the time complexity is O(n^2).

```python
%%time
l1 = list(map(int,input().split()))
n = len(l1)
if n%2 == 0:
    a = n//2-1
else:
    a = n//2
for i in range(a):
    max_element = max(l1)
    l1.remove(max_element)
if n%2 == 0:
    max_1 = max(l1)
    l1.remove(max_1)
    max_2 = max(l1)
    print((max_1+max_2)/2)
else:
    print(max(l1))
```

A better method would be to use the sorted function and then finding out the median based on if the length of the list is odd or even.

```python
%%time
l1 = list(map(int,input().split()))
n = len(l1)
l1.sort()
# print(n)
# print(l1)
if n%2 == 0:
    # print(n//2)
    median = (l1[n//2-1] + l1[n//2])/2
else:
    median = l1[(n+1)//2]
print(median)
```

# Q4. What is the gradient of the following function with respect to x and y?

$x^2y + y^3sin(x)$

$f(x,y) = x^2y + y^3sin(x)$ 

$\frac{\partial{f(x,y)}}{\partial{x}} = 2xy + y^3cos(x)$

$\frac{\partial{f(x,y)}}{\partial{y}} = x^2 + 3y^2sin(x)$

$Therefore,\: the\: gradient\: is\: 2xy + y^3cos(x)\hat{i} + x^2 + 3y^2sin(x)\hat{j}.$

# Q5. Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y

```python
import jax
import jax.numpy as jnp

def f(x, y):
     return x**2 * y + y**3 * jnp.sin(x)

def gradient(x, y):
    df_dx = 2 * x * y + y**3 * jnp.cos(x)
    df_dy = x**2 + 3 * y**2 * jnp.sin(x)
    return jnp.array([df_dx, df_dy])     

grad_f = jax.grad(f,(0,1))

x = float(input('Enter the value of x:'))
y = float(input('Enter the value of y:'))

analyt_grad = jnp.array(grad_f(x,y))
grad = gradient(x, y)
print(f'The gradient evaluated by my method is {grad}')
print(f'The analytical gradient is {analyt_grad}')
```

```python
import sympy as sp
import jax
import jax.numpy as jnp

x,y = sp.symbols('x y')

f_x_y = x**2 * y + y**3 * sp.sin(x)

df_x_y_dx = sp.diff(f_x_y, x)
df_x_y_dy = sp.diff(f_x_y, y)

# We need to lambdify the sympy expressions to evaluate them numerically
df_x_y_dx_func = sp.lambdify((x, y), df_x_y_dx, 'numpy')
df_x_y_dy_func = sp.lambdify((x, y), df_x_y_dy, 'numpy')

def gradient2(x_value, y_value):
    df_x_y_dx_value = df_x_y_dx_func(x_value, y_value)
    df_x_y_dy_value = df_x_y_dy_func(x_value, y_value)
    return jnp.array([df_x_y_dx_value, df_x_y_dy_value])

x_val = float(input('Enter the value of x:'))
y_val = float(input('Enter the value of y:'))

grad = gradient2(x_val, y_val)
print(grad)
```

# Q9. Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.

# Q10. Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as “a”, “b”, “c”, “d”, “e”. Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import numpy as np
import pandas as pd

matrix = np.random.uniform(1,2,(20,5))

df = pd.DataFrame(matrix, columns =['a','b','c','d','e'])

max_std_dev_col = df.std().idxmax()
max_std_dev = df[max_std_dev_col].std()
#print(df[['a','b','c','d','e']].std())
# idxmax is a very useful function as it saves a lot of time

mean_all = df.mean(axis=1)
# print(mean_all)
min_mean_row = df.mean(axis=1).idxmin()
min_mean = min(mean_all)

print(f"Column with highest standard deviation is {max_std_dev_col} and it's value is {max_std_dev}")
print(f"Row with the lowest mean is {min_mean_row} and it's value is {min_mean}")
```
